Start cleaning data:-

1. Correct data types:

```python

df['Pclass']=df['Pclass'].astype('category')
df['Pclass']= df['Pclass'].cat.rename_categories({1:'1st',2:'2nd',3:'3rd'})
df['Sex']=df['Sex'].astype('category')
df['Embarked']=df['Embarked'].astype('category')
#check after adjusting:
df.info()
df.head()

```

2. Round up age values:

```python
df['Age']= df['Age'].round(2)
df['Age'].unique()
```

3. check for duplicates:

```python
df.duplicated().sum
```

4. filling the two missed embarked values with the most frequent value (mfv)


```python
mfv=df['Embarked'].mode()[0]
df['Embarked'].fillna(mfv,inplace=True)
```

Working on Age:
1- Understanding the data and its defects

```python
age_stats = df.groupby('Pclass')['Age'].agg(['min', 'max', 'mean', 'median', 'count', 'size'])
age_stats
```

```python
age_stats['range']=age_stats['max']-age_stats['min']
age_stats
```

```python
#calculating total missing ages
total_missing_ages =df['Age'].isna().sum()
total_missing_ages
```

```python
#calculate total missing Ages (tma) for each class of passengers
tma=df.groupby('Pclass')['Age'].apply(lambda x: x.isna().sum())
tma
#calculate the percentage of missing ages for each
total_age_by_class=df.groupby(['Pclass'])['Age'].size()
total_age_by_class
tma_percentage= (tma/total_age_by_class)*100
tma_percentage
```

5. Filling the missing ages by the mean value of ages in each class:-


```python
df['Age']=df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.mean()))
df['Age']
```

All age entries are filled


* Exploratory Data Analysis (EDA)
* understanding the relation between survival rates and other factors.


```python
total_survived=df['Survived'].sum()
percentage=(total_survived/df['Survived'].count())*100
percentage
```

1. Understanding cabin and class factors:

```python
# Extract deck letter
def extract_deck(cabin):
    if pd.notna(cabin):
        cabin_str=str(cabin).strip()
        deck=cabin_str[0]
        if deck in ['A', 'B', 'C', 'D', 'E', 'F', 'G']:
            return deck
        else:
            return 'nonfamiliar deck'
    else:
        return 'missing'
df['Deck']=df['Cabin'].apply(extract_deck)
#group deck and class
deck_class=pd.crosstab(df['Pclass'],df['Deck'])
deck_class
```

```python
deck_class_survival = df.groupby(['Deck', 'Pclass'])['Survived'].mean() * 100
print("\n=== Survival Rate by Deck AND Passenger Class ===")
print(deck_class_survival.sort_values(ascending=False))
```

2. Understanding gender factor:

```python
sex_survival = df.groupby('Sex')['Survived'].mean() * 100
sex_survival
```

Females had ~74% survival, males only ~19%.

3. Understanding age factor:

```python
def age_group(Age):
     if 12 >= Age:
        return "Child (0-12)"
     elif 18 >= Age:
        return "Teen (13-18)"
     elif 50 >= Age:
        return "Adult (19-50)"
     else:
        return 'Senior (51+)'

df['Age_group'] = df['Age'].apply(age_group)
age_survival = df.groupby('Age_group')['Survived'].mean() * 100
print(age_survival)
```

4. Understanding family size factor:

```python
# Calculate family size
df['Family_size'] = df['SibSp'] + df['Parch']

# categorize family size
def family_category(size):
    if size == 0:
        return 'Alone'
    elif size <= 3:
        return 'Small Family (1-3)'
    else:
        return 'Large Family (4+)'

df['Family_category'] = df['Family_size'].apply(family_category)
family_survival = df.groupby('Family_category')['Survived'].mean() * 100
family_survival
```
